#**Read and Process Data**

```
def pair_to_vector(idx, length):
    res = [0] * length
    for i in idx:
        res[i] = 1
    return res
```

```
FILENAMES = ["journal_of_international_economics_final.xlsx", "the_world_economy_final.xlsx", "journal_of_development_economics_final.xlsx", "world_development_final.xlsx"]

authors = []
papers = []
author_paper_pair = []

for file in FILENAMES:
    # open workbook
    wb = openpyxl.load_workbook(file)
    ws = wb.active
    MAXROW = ws.max_row
    MAXCOL = ws.max_column
    curr = 0
    curr_idx = -1

    for r in ws.rows:
        # ignore header
        if(r[0].row == 1):
            continue
        # halt loop
        if(r[0].row > MAXROW):
            break
        # update cursor
        if(isinstance(r[2].value, str)):
            curr = r[2].value
            papers.append(curr)
            curr_idx += 1
        # row corresponds to a keyword
        if(isinstance(r[4].value, str)):
            # duplicate exists
            if(r[4].value in authors):
                author_paper_pair[authors.index(r[4].value].append(curr_idx)
            # duplicate doesn't exist
            else:
                authors.append(r[4].value)
                author_paper_pair.append([curr_idx])

# encode keyword-paper pair to vector
author_vector = []
for i in range(len(author_paper_pair)):
    author_vector.append(pair_to_vector(author_paper_pair[i], len(papers)))

wb.close()
```

#**Write Data to CSV**

```
gc.collect()
```

```
df = pd.DataFrame(author_vector)
df = df.transpose()
df.index = papers
df.columns = authors
```

```
df
```

```
df.to_csv("final_paper-author_mat.csv", mode='w')
```

#**Sort by Frequency**

```
freq = df.sum(axis=0).tolist()
freq_mat = []
for i in range(len(freq)):
    freq_mat.append([authors[i], freq[i]])
freq_mat.sort(key=lambda x: x[1])
freq_mat.reverse()
pd.DataFrame(freq_mat).to_csv("final_paper-author_freq.csv", mode="w", header=False, index=False)
```
