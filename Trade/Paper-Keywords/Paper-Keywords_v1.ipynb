#**Import Module**

#**Open Workbook**

```
wb = openpyxl.load_workbook('final.xlsx')
```

#**Read and Process Data**

```
def pair_to_vector(idx, length):
    res = [0] * length
    for i in idx:
        res[i] = 1
    return res
```

```
ws = wb.active
MAXROW = ws.max_row
MAXCOL = ws.max_column

keywords = []
papers = []
keyword_paper_pair = []
curr = 0
curr_idx = -1

for r in ws.rows:
    # ignore header
    if(r[0].row == 1):
        continue
    # halt loop
    if(r[0].row > MAXROW):
        break
    # update cursor
    if(isinstance(r[2].value, str)):
        curr = r[2].value
        papers.append(curr)
        curr_idx += 1
    # row corresponds to a keyword
    if(isinstance(r[3].value, str)):
        # duplicate exists
        if(r[3].value.lower() in keywords):
            keyword_paper_pair[keywords.index(r[3].value.lower())].append(curr_idx)
        # duplicate doesn't exist
        else:
            keywords.append(r[3].value.lower())
            keyword_paper_pair.append([curr_idx])

# encode keyword-paper pair to vector
keyword_vector = []
for i in range(len(keyword_paper_pair)):
    keyword_vector.append(pair_to_vector(keyword_paper_pair[i], curr_idx + 1))
```

#**Write Data**

```
wb.create_sheet("analysis")
ws = wb.get_sheet_by_name("analysis")

for i in range(len(papers)):
    ws.cell(row = i + 2, column = 1).value = papers[i]
    if i % 500 == 0:
        print(i / 500 * 100)
        gc.collect()
for j in range(len(keyword_vector)):
    ws.cell(row = 1, column = j + 2).value = keywords[j]
    for i in range(len(keyword_vector[j])):
        ws.cell(row = i + 2, column = j + 2).value = keyword_vector[j][i]
    if j % 500 == 0:
        print(j / 500 * 100)
        gc.collect()
```
