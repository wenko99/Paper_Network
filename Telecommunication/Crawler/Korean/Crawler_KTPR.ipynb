#**Fetch all Volume Links**

```
MAINURL = "http://kiss.kstudy.com/journal/journal-view.asp?key1=2256&key2=3041"
volUrls = []

# access main page
wd.get(MAINURL)
time.sleep(0.5)

try:
    # move tab
    jour_tab = wd.find_element_by_xpath("//a[@href='#jour2']")
    ActionChains(wd).move_to_element(jour_tab).click().perform()     
    time.sleep(0.5)

    # click all 더보기 buttons
    while True:
        try:
            more_btn = wd.find_element_by_xpath("//input[@id='btnAddView']")
            if "none" in more_btn.get_attribute("style"):
                break
            else:
                ActionChains(wd).move_to_element(more_btn).click().perform()
                time.sleep(0.5)
                continue
        except NoSuchElementException:
            break
    
    # fetch all volume urls
    rows = wd.find_elements_by_xpath("//div[@id='jour2']/table/tbody/tr")
    for row in rows:
        yrs = row.find_elements_by_xpath("./th")
        vols = row.find_elements_by_xpath("./td")
        for i in range(len(yrs)):
            for vol in vols[i].find_elements_by_xpath("./p/a[@href]"):
                volUrls.append((yrs[i].get_attribute("innerText")[:4], vol.get_attribute("href")))
except NoSuchElementException:
    print("NO DIV TAG")

print(volUrls)
```

#**Fetch and Write Data from each Volumes**

```
wb = openpyxl.Workbook()
ws = wb.active
ws.title = "kwd"
```

```
TARGETPATH = "./Data/raw/Korean/"
```

```
 r = 2

 # write header
 ws.cell(row = 1, column = 1).value = 'title'
 ws.cell(row = 1, column = 2).value = 'journal'
 ws.cell(row = 1, column = 3).value = 'year'
 ws.cell(row = 1, column = 4).value = 'keywords'
 wb.create_sheet("auth")
 ws = wb["auth"]
 ws.cell(row = 1, column = 1).value = 'title'
 ws.cell(row = 1, column = 2).value = 'authors'
 ws = wb["kwd"]

 for (yr, volurl) in volUrls:
    # access volume page
    wd.get(volurl)
    time.sleep(1)

    # fetch volume no. from url
    print("Reading Volume from year, " + str(yr) + " : "  + volurl)

    # fetch article urls
    articleUrls = []
    a_tags = wd.find_elements_by_xpath("//div[@class='thesis-info']/h5/a[@href]")
    for tag in a_tags:
        articleUrls.append(tag.get_attribute("href"))

    for articleurl in articleUrls:
        # access article page
        wd.get(articleurl)
        time.sleep(0.8)
        #print("Reading article, " + articleurl)

        # fetch title
        try:
            title = wd.find_element_by_xpath("//section[@class='pub-info']/h3").get_attribute("innerText")
        except NoSuchElementException:
            continue
        #print(title)
        
        # fetch authors
        authors = []
        try:
            authors = wd.find_element_by_xpath("//div[@class='writers']").get_attribute("innerText").split(", ")
            while ' ' in authors: 
                authors.remove(' ')
            for i in range(len(authors)):
                if "(" in authors[i]:
                    authors[i] = authors[i][:authors[i].index("(") - 1]
        except NoSuchElementException:
            continue
        if(len(authors) == 0):
            continue
        #print(authors)

        # fetch keywords
        keywords = []
        for kwd in wd.find_elements_by_xpath("//div[@class='key-words is1']/span"):
            keywords.append(kwd.get_attribute("innerText"))
        while ' ' in keywords:
            keywords.remove(' ')
        hanguel_idx = []
        for i in range(len(keywords)):
            if len(re.compile("[가-힣]+").findall(keywords[i])) > 0:
                hanguel_idx.append(i)
        hanguel_idx.reverse()
        for i in hanguel_idx:
            del keywords[i]
        if(len(keywords) == 0):
            continue
        #print(keywords)

        # fetch month
        '''
        try:
            month = wd.find_element_by_xpath("//label[contains(text(), '발행년월')]").find_element_by_xpath("..").get_attribute("innerText")
            month = month[month.index("년") + 1:]
            month = month[month.index("년") + 2:-1]
        except ValueError:
            continue
        except NoSuchElementException:
            continue
        #print(month)
        '''

        # write data to excel file
        print("Writing, " + title)
        ws = wb["kwd"]
        ws.cell(row = r, column = 1).value = title
        ws.cell(row = r, column = 2).value = "Korean Telecommunications Policy Review"
        ws.cell(row = r, column = 3).value = yr
        for j in range(len(keywords)):
            ws.cell(row = r, column = 4 + j).value = keywords[j]
        ws = wb["auth"]
        ws.cell(row = r, column = 1).value = title
        for j in range(len(authors)):
            ws.cell(row = r, column = 2 + j).value = authors[j]
        r += 1
    
    # backup for every ten volumes
    if(volUrls.index((yr, volurl)) % 20 == 0):
        print("Saving backup file")
        wb.save(filename = TARGETPATH + "korean_telecommunications_policy_review_backup.xlsx")
        time.sleep(20)
        #files.download(TARGETPATH + "korean_telecommunications_policy_review_backup.xlsx")
```
